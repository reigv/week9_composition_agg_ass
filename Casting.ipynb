## avoid c-style casting in c++

- danger!! poiter/int or unrelates ptr conversions
- use c++ style casts instead
- static_cast, dynamic_cast, const_cast, reinterpret_cast

- oh it can remove constness too, be careful, without any warning


```python
#include <iostream>

int main()
{

    int i = 10;

    int *p = (int *)i; // casting an integer as an address

    std::cout << "Raw value: " << i << std::endl;

    // this line will lead to a crash
    std::cout << "Pointer value: " << *p << std::endl;

    return 0;
}
```

### static_casting
- used for conversions that are known to be safe at compile time
- similar to C-style casting but more restrictive and safer
