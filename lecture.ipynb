# Compositions, Aggregations, Associations, and casting

## Intro to Relationships
- inheritance is 1 type of **relationship* between diff classes
- it is a **is-a** relationship -- meaning that the child class is a type of the parent class
---> new: **has-a** relationship -- meaning that one class contains another class as a part of itself
- 3 types of has-a relationships:
    1. **composition** --- strong relate, where the contained class cannNOT exist wihtout the container class
    2. ** aggregation** ---- weaker relationship, where the contained cclass can exist without the container
    3. **association** -- weakest, where classes are aware of each other but do not contain eachother

Compositions: The composer object owns and controls the lifecycle of the component object

Aggregation: The aggregator has an instance of another class, but does not control its lifecycle

Association: One class accesses or uses another class

## Composition
- strong relationship between classes
---> ONE obj ----> creating && destroying its component objs

- frequently used with member variables 
- ex: Car class has Engine class as a component. The Car object creates and destroys the Engine object as part of its lifecycle.

- can also be used for **poiters** to indicate ownership and control over the lifecycle of the component object
- ex: Car class has a pointer to an Engine object. The Car class is responsible for creating and destroying the Engine object, indicating a strong relationship between the two classes.


```python
#include <iostream>
#include <string>

using namespace std;

class Engine
{
    int hp_;
    string type_;

public:
    Engine(int hp, string type) : hp_(hp), type_(type)
    {
        cout << "Constructor for Engine called." << endl;
    }
    ~Engine()
    {
        cout << "Destructor for Engine called." << endl;
    }
    void display()
    {
        cout << "Engine: " << type_
             << ", HP: " << hp_ << endl;
    }
};

class Car
{
    Engine engine_; // composition: Car has an Engine
    // direct data members of Car
    //this is not a pointer, so the Engine object is created and destroyed as part of the Car object's lifecycle

    string maker_;
    string model_;
    int doors_;

public:
    Car(string maker, string model, int doors, int hp,
        string type) : engine_(hp, type), maker_(maker),
                       model_(model), doors_(doors)
    {
        cout << "Constructor for Car called." << endl;
    }
    ~Car()
    {
        cout << "Destructor for Car called." << endl;
    }
    void display()
    {
        cout << "Maker: " << maker_ << ", Model: " << model_
             << ", Number of doors: " << doors_ << endl;
        engine_.display();
    }
};

int main()
{

    Car *corolla = new Car("Toyota", "Corolla", 4, 170, "V4");

    corolla->display();

    delete corolla;

    return 0;
}
```

## Aggregation

-weaker
- still the has-a, but the contained class can exist independently of the container class
- ex: University class has a list of Student objects. The University class can exist without the Student objects, and the Student objects can exist without the University class.

--- POINTERS ||| arr POINTers

```python
#include <iostream>
#include <string>

using namespace std;

class Player
{
private:
    string name_;
    int number_;

public:
    Player(string name, int number) : name_(name), number_(number)
    {
        cout << "Constructor for player called." << endl;
    }
    ~Player()
    {
        cout << "Destructor for player called." << endl;
    }
    void display() const
    {
        cout << "Name: " << name_
             << ", Number: " << number_ << endl;
    }
};

class Team
{
private:
    string name_;
    Player *players_;
    int num_players_;

public:
    Team(string name, int num, Player *players)
        : name_(name), num_players_(num), players_(players)
    {
        cout << "Constructor for team called." << endl;
    }
    ~Team()
    {
        cout << "Destructor for team called." << endl;
    }
    void display() const
    {
        for (int i = 0; i < num_players_; i++)
        {
            players_[i].display();
        }
    }
};

int main()
{

    Player players[3] = {
        Player("Dembele", 10),
        Player("Marquinhos", 4),
        Player("Hakimi", 2)};

    Team PSG("PSG", 3, players);

    PSG.display();
    delete PSG; // this will cause an error because we are trying to delete a stack-allocated object, not a heap-allocated object
    
    return 0;


}
```

### ASSOCIATION
- weakest relationship between classes
- otherwise known as a "uses-a" relationship ---- unrelated objects that interact with each other
- ex: Teacher class and Student class. A Teacher object can interact with a Student object, but they do not contain each other or have a strong relationship.

```python
#include <iostream>
#include <string>

using namespace std;

class Course; // forward declaration

class Student
{
    string name_;
    Course *courses_[5];
    int courseCount_;

public:
    Student(const string &name) : name_(name), courseCount_(0) {}

    const string &name() const { return name_; }

    void addCourse(Course *course)
    {
        if (courseCount_ < 5)
        {
            courses_[courseCount_++] = course;
        }
    }

    void printCourses() const;
};

class Course
{
    string name_;
    string code_;
    Student *students_[10];
    int studentCount_;

public:
    Course(const string &name, const string &code)
        : name_(name), code_(code), studentCount_(0) {}

    const string &name() const
    {
        return name_;
    }

    const string &code() const
    {
        return code_;
    }

    void addStudent(Student *student)
    {
        if (studentCount_ < 10)
        {
            students_[studentCount_++] = student;
        }
    }

    void printStudents() const
    {
        cout << code_ << ": " << name_ << " roster: ";
        for (int i = 0; i < studentCount_; i++)
        {
            cout << students_[i]->name();
            if (i + 1 < studentCount_)
                cout << ", ";
        }
        cout << endl;
    }
};

void Student::printCourses() const
{
    cout << name_ << " is taking: ";
    for (int i = 0; i < courseCount_; i++)
    {
        cout << courses_[i]->code()
             << ": " << courses_[i]->name();
        if (i + 1 < courseCount_)
            cout << ", ";
    }
    cout << endl;
}

void enroll(Student &student, Course &course)
{
    course.addStudent(&student);
    student.addCourse(&course);
}

int main()
{
    Student alice("Alice"), bob("Bob"),
        john("John"), kevin("Kevin");
    Course sep200("Object Oriented Programming", "SEP200"),
        sem305("Discrete Mathematics", "SEM305");

    enroll(alice, sep200);
    enroll(alice, sem305);
    enroll(bob, sep200);
    enroll(john, sep200);
    enroll(kevin, sem305);

    sep200.printStudents();
    sem305.printStudents();

    alice.printCourses();

    return 0;
}
```
